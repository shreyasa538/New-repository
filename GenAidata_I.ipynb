```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/static.csv")
data.drop(data.tail(1).index,inplace=True)
data['period_end']=pd.to_datetime(data['period_end'])
```

```python
plt.figure(figsize=(15,6))
sns.set_theme(style="whitegrid")
plt.title("number of followers gained every month")
sns.barplot(x="period_end",y="followers_gained",data=data)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Plot 2: Total followers

plt.figure(figsize=(15, 6))

sns.set_theme(style="whitegrid")

plt.title("Total Followers At The End of Every Month")

sns.barplot(x="period_end", y="followers_total", data=data)

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15, 6))

sns.set_theme(style="whitegrid")

plt.title("Total Views Every Month")

sns.barplot(x="period_end", y="views", data=data)

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
from autots import AutoTS

#Make sure only the necessary data is passed

ts_data = data[['period_end', 'followers_gained']].copy()

model = AutoTS(
forecast_length=2,
frequency='infer',
ensemble='simple',
min_allowed_train_percent=0.3

)

model = model.fit(ts_data, date_col='period_end', value_col='followers_gained', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
